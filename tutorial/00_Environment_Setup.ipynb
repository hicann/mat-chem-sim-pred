## 仓库结构验证

确认仓库目录结构完整，所有主要模块都在正确位置。

```python
import os
repo_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
print(f"仓库根目录: {repo_root}")

expected_dirs = [
    'simulation/AI4MD',
    'simulation/AI4PDE',
    'simulation/MaterialPropertyPrediction',
    'prediction/TabularData',
    'prediction/TimeSeries',
    'prediction/SmallData',
    'template',
]

for d in expected_dirs:
    full = os.path.join(repo_root, d)
    exists = os.path.isdir(full)
    print(f"  {'✅' if exists else '❌'} {d}")
```

```python
print("\n环境检查完成，运行简单矩阵乘法测试...")
try:
    import torch
    a = torch.randn(100, 100)
    b = torch.randn(100, 100)
    c = a @ b
    print(f"PyTorch 矩阵乘法测试: ✅ 正常 (输出形状 {c.shape})")
except Exception as e:
    print(f"PyTorch 测试失败: {e}")
```
